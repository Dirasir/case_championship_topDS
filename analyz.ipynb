```python
import pandas as pd
```

```python
NAME = "oil"
```

```python
df = pd.read_json(f"{NAME}.json", lines=True)
df = df.rename(columns={'rating': 'rating_of_organization'})
df.head(10)
```

```python
print(len(df))
```

```python
df_expanded = df.explode('reviews').reset_index(drop=True)
review_fields = pd.json_normalize(df_expanded['reviews'].tolist())
df_final = pd.concat([
    df_expanded.drop('reviews', axis=1).reset_index(drop=True),
    review_fields
], axis=1)
```

```python
df_final.drop(["categories","goods","opening_hours","phones"], inplace=True, axis=1)
df_final.head()
```

```python
df_final.to_csv(f"exploded_{NAME}.csv", index=False)
```

## Short Variant

```python
df.drop(["categories", "goods", "opening_hours", "phones", "city"], inplace=True, axis=1)
df.head()
```
